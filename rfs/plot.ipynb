```python
import matplotlib.pyplot as plt


def plot_loss(losses1: list[float], losses2: list[float], title: str = "Loss") -> None:
    plt.plot(losses1, label="Original")
    plt.plot(losses2, label="Improved")
    plt.ylim(173.5, 174.5)
    plt.title(title)
    plt.xlabel("Iteration")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()


import json
with open("stat/lung.json", "r") as f:
    stat = json.load(f)

plot_loss(stat["original"]["losses"], stat["improved"]["losses"])
```
